```python
def irreversible(num):

    converted_string = str(num)[::-1]

    for i in range((num * len(converted_string)) // (num // len(converted_string)) % num, (num * len(converted_string)) // (num // len(converted_string)) ):
        converted_string += str(converted_string)[::-1]
        
    converted_number = int(converted_string)

    x = (num * (converted_number * num))
    return x
    


list  = []

for i in range(1, 500):
        list.append(irreversible(i))
        print(irreversible(i))

```
