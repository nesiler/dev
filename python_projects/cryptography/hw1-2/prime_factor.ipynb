```python
import time, math, itertools

def is_prime(n):
    if n <= 1:
        return False
    for i in range(2,n):
        if n % i == 0:
            return False
    return True

def prime_factor(m):
    ls = []
    for i in itertools.chain([2], itertools.count(3, 1)):
        if (m <= 1):
            break
        while (m % i == 0):
            m //= i
            yield i
            ls.append(i)
    return ls

def is_primeS(m, n):
    m_factors = prime_factor(m)
    n_factors = prime_factor(n)

    for i in m_factors:
        for j in n_factors:
            if (i == j):
                return False
    
    return True
```

```python
t0 = time.time()

prime_numbers = []

for i in range(100,1000):
    if(is_prime(i)):
        prime_numbers.append(i)
        print(i)
     
t1 = time.time()
print("Time required :", t1 - t0)
```
